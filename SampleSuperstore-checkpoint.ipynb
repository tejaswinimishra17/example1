## Project Overview:

This data analysis project aims to explore and gain insights from the 'SampleSuperstore' dataset, which contains information about sales and operations of a retail superstore. The project focuses on understanding various aspects of the business, including shipping modes, customer segments, geographical performance, product categories, discounts, quantity sold, profitability, and customer behavior. By analyzing these aspects, the project seeks to identify opportunities for improving sales, profit margins, and overall business efficiency.

## Problem Statement:

The retail superstore faces several challenges and opportunities for optimization. The primary objective of this project is to answer critical business questions and provide actionable insights to address the following key issues:

1. **Shipping Mode Analysis:**
   - Identify the most commonly used shipping modes to streamline delivery operations.
   - Investigate whether there is a correlation between shipping modes and profit margins to optimize shipping strategies.

2. **Customer Segment Analysis:**
   - Analyze the contribution of different customer segments (e.g., corporate, consumer) to sales and profit to tailor marketing and service strategies.
   - Determine which customer segments are the most profitable to focus on high-value customers.

3. **Geographic Analysis:**
   - Understand the distribution of sales and profit by country, state, city, and region to identify top-performing and underperforming areas.
   - Investigate whether specific regions or states consistently perform better or worse in terms of sales and profit.

4. **Category and Sub-Category Analysis:**
   - Identify the product categories and sub-categories with the highest and lowest sales and profit to optimize product offerings.
   - Find specific sub-categories showing potential for growth or optimization to expand the product portfolio.

5. **Discount Impact Analysis:**
   - Analyze how discounts affect sales and profit to make informed decisions regarding discount strategies.
   - Determine optimal discount ranges for maximizing profit while maintaining competitiveness.

6. **Quantity Analysis:**
   - Explore the relationship between quantity sold and profit to optimize inventory management and pricing strategies.
   - Identify products or categories that sell in larger quantities to adjust marketing and stock levels.

7. **Profitability by Category and Sub-Category:**
   - Determine the profitability of different product categories and sub-categories to allocate resources effectively.
   - Evaluate whether specific sub-categories should be expanded or discontinued based on profitability.

8. **Segmented Analysis:**
   - Perform further analysis of customer segments to understand high-value customer behaviors and preferences.
   - Investigate how different segments respond to discounts, promotions, and product categories for targeted marketing efforts.

By addressing these questions and challenges, the project aims to provide data-driven recommendations and insights to enhance the superstore's overall performance, increase profitability, and better serve its customers.

## Exploratory Data Analysis and Data Vizualisation

```python
# Summary Statistics
summary_stats = df[['Sales', 'Profit', 'Quantity', 'Discount']].describe()
print("\033[1m Summary Statistics:\033[0m")
print(tabulate(summary_stats, headers='keys', tablefmt='grid'))
```

## 1. Shipping Mode Analysis:

### 1.1 What are the most commonly used shipping modes?

```python
# Count the occurrences of each shipping mode
shipping_mode_counts = df['Ship Mode'].value_counts()

# Create a bar chart to visualize shipping mode counts
plt.figure(figsize=(10, 6))
plt.bar(shipping_mode_counts.index, shipping_mode_counts.values, color='lightblue')
plt.title('Most Commonly Used Shipping Modes')
plt.xlabel('Shipping Mode')
plt.ylabel('Count')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the bar chart
plt.tight_layout()
plt.show()

# Display the most commonly used shipping modes
most_common_shipping_modes = shipping_mode_counts.head()
print("Most commonly used shipping modes:")
print(most_common_shipping_modes)
```

### 1.2 Is there a correlation between shipping mode and profit margin?

```python
# Calculate profit margin (Profit / Sales)
df['Profit Margin'] = df['Profit'] / df['Sales']

# Perform one-way ANOVA to test for correlation between shipping mode and profit margin
shipping_modes = df['Ship Mode'].unique()
anova_results = {}

for mode in shipping_modes:
    group_data = df[df['Ship Mode'] == mode]['Profit Margin']
    anova_results[mode] = group_data

f_statistic, p_value = stats.f_oneway(*anova_results.values())

# Define the significance level
alpha = 0.05

# Determine if there is a significant correlation
if p_value < alpha:
    print("There is a significant correlation between shipping mode and profit margin.")
else:
    print("There is no significant correlation between shipping mode and profit margin.")

# Create a box plot to visualize profit margins for each shipping mode
plt.figure(figsize=(10, 6))
sns.barplot(x='Ship Mode', y='Profit Margin', data=df, palette='Set3')
plt.title('Profit Margin by Shipping Mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Profit Margin')

# Show the box plot
plt.tight_layout()
plt.show()
```

## 2. Customer Segment Analysis:

### 2.1 How do different customer segments (e.g., corporate, consumer) contribute to sales and profit?

```python
# Group the data by 'Segment' and calculate total sales and profit for each segment
segment_sales_profit = df.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()

# Create a grouped bar chart to visualize sales and profit by segment
plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(segment_sales_profit))
plt.bar(index, segment_sales_profit['Sales'], bar_width, label='Sales', color='skyblue', edgecolor='black')
plt.bar([i + bar_width for i in index], segment_sales_profit['Profit'], bar_width, label='Profit', color='salmon', edgecolor='black')
plt.xlabel('Segment')
plt.ylabel('Amount (USD)')
plt.title('Sales and Profit by Customer Segment')
plt.xticks([i + bar_width/2 for i in index], segment_sales_profit['Segment'])
plt.legend()

# Show the grouped bar chart
plt.tight_layout()
plt.show()

# Print the contribution of each segment to sales and profit
print("Contribution of different customer segments to Sales and Profit:")
print(segment_sales_profit)
```

### 2.2 Which customer segments are the most profitable?

```python
# Calculate profit per order
df['Profit per Order'] = df['Profit'] / df['Quantity']

# Group the data by 'Segment' and calculate the average profit per order for each segment
segment_avg_profit = df.groupby('Segment')['Profit per Order'].mean().reset_index()

# Find the most profitable segments
most_profitable_segments = segment_avg_profit[segment_avg_profit['Profit per Order'] == segment_avg_profit['Profit per Order'].max()]

# Create a bar chart to visualize the average profit per order by segment
plt.figure(figsize=(10, 6))
plt.bar(segment_avg_profit['Segment'], segment_avg_profit['Profit per Order'], color='skyblue', edgecolor='black')
plt.title('Average Profit per Order by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Average Profit per Order (USD)')
plt.xticks(rotation=45)

# Show the bar chart
plt.tight_layout()
plt.show()

# Print the most profitable segments
print("The most profitable customer segment(s):")
print(most_profitable_segments)
```

```python
# Group the data by 'Country', 'State', 'City', and 'Region' and calculate total sales and profit for each group
country_sales_profit = df.groupby('Country')[['Sales', 'Profit']].sum().reset_index()
state_sales_profit = df.groupby('State')[['Sales', 'Profit']].sum().reset_index()
city_sales_profit = df.groupby('City')[['Sales', 'Profit']].sum().reset_index()
region_sales_profit = df.groupby('Region')[['Sales', 'Profit']].sum().reset_index()

# Create subplots for effective visualization
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Subplot 1: Distribution of Sales and Profit by Country
axs[0, 0].bar(country_sales_profit['Country'], country_sales_profit['Sales'], label='Sales', color='skyblue', edgecolor='black')
axs[0, 0].bar(country_sales_profit['Country'], country_sales_profit['Profit'], label='Profit', color='salmon', edgecolor='black', alpha=0.7)
axs[0, 0].set_title('Sales and Profit by Country')
axs[0, 0].set_xlabel('Country')
axs[0, 0].set_ylabel('Amount (USD)')
axs[0, 0].tick_params(axis='x', rotation=90)
axs[0, 0].legend()

# Subplot 2: Distribution of Sales and Profit by State
axs[0, 1].bar(state_sales_profit['State'], state_sales_profit['Sales'], label='Sales', color='skyblue', edgecolor='black')
axs[0, 1].bar(state_sales_profit['State'], state_sales_profit['Profit'], label='Profit', color='salmon', edgecolor='black', alpha=0.7)
axs[0, 1].set_title('Sales and Profit by State')
axs[0, 1].set_xlabel('State')
axs[0, 1].set_ylabel('Amount (USD)')
axs[0, 1].tick_params(axis='x', rotation=90)
axs[0, 1].legend()

# Subplot 3: Distribution of Sales and Profit by City (Displaying the top 10 cities)
top_10_cities = city_sales_profit.sort_values(by='Sales', ascending=False).head(10)
axs[1, 0].bar(top_10_cities['City'], top_10_cities['Sales'], label='Sales', color='skyblue', edgecolor='black')
axs[1, 0].bar(top_10_cities['City'], top_10_cities['Profit'], label='Profit', color='salmon', edgecolor='black', alpha=0.7)
axs[1, 0].set_title('Sales and Profit by City (Top 10)')
axs[1, 0].set_xlabel('City')
axs[1, 0].set_ylabel('Amount (USD)')
axs[1, 0].tick_params(axis='x', rotation=45)
axs[1, 0].legend()

# Subplot 4: Distribution of Sales and Profit by Region
axs[1, 1].bar(region_sales_profit['Region'], region_sales_profit['Sales'], label='Sales', color='skyblue', edgecolor='black')
axs[1, 1].bar(region_sales_profit['Region'], region_sales_profit['Profit'], label='Profit', color='salmon', edgecolor='black', alpha=0.7)
axs[1, 1].set_title('Sales and Profit by Region')
axs[1, 1].set_xlabel('Region')
axs[1, 1].set_ylabel('Amount (USD)')
axs[1, 1].tick_params(axis='x', rotation=0)
axs[1, 1].legend()

# Adjust layout and show subplots
plt.tight_layout()
plt.show()

# Print the distribution of sales and profit by country, state, city, and region
print("\033[1m1.Distribution of Sales and Profit by Country:\033[0m")
print(country_sales_profit.head())
print("\n\033[1m2.Distribution of Sales and Profit by State:\033[0m")
print(state_sales_profit.head())
print("\n\033[1m3.Distribution of Sales and Profit by City:\033[0m")
print(city_sales_profit.head())
print("\n\033[1m4. Distribution of Sales and Profit by Region:\033[0m")
print(region_sales_profit)
```

### 3.2 Are regions or states consistently performing better or worse in terms of sales and profit?

```python
# Calculate profit margin (Profit / Sales)
df['Profit Margin'] = df['Profit'] / df['Sales']

# Group the data by 'Region' and 'State' and calculate average sales and profit for each
region_avg_sales_profit = df.groupby(['Region'])[['Sales', 'Profit']].mean().reset_index()
state_avg_sales_profit = df.groupby('State')[['Sales', 'Profit']].mean().reset_index()

# Sort the states by average profit in descending order
state_avg_sales_profit = state_avg_sales_profit.sort_values(by='Profit', ascending=False)

# Create subplots for effective visualization
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Subplot 1: Average Sales and Profit by Region
sns.barplot(x='Region', y='Sales', data=region_avg_sales_profit, ax=axs[0], color='skyblue', label='Sales')
sns.barplot(x='Region', y='Profit', data=region_avg_sales_profit, ax=axs[0], color='salmon', label='Profit')
axs[0].set_title('Average Sales and Profit by Region')
axs[0].set_xlabel('Region')
axs[0].set_ylabel('Amount (USD)')
axs[0].legend()

# Subplot 2: Average Profit by State (Top 10 States)
sns.barplot(x='Profit', y='State', data=state_avg_sales_profit.head(10), ax=axs[1], palette='Set3', ci=None)
axs[1].set_title('Average Profit by State (Top 10)')
axs[1].set_xlabel('Average Profit')
axs[1].set_ylabel('State')

# Adjust layout and show subplots
plt.tight_layout()
plt.show()

# Output the top-performing states
top_states = state_avg_sales_profit.head(10)
print("\033[1mTop-Performing States (by Average Profit):\033[0m")
print(top_states)
```

## 5. Discount Impact Analysis:

### 5.1 How does the level of discount offered affect sales and profit?

```python
# Calculate profit margin (Profit / Sales)
df['Profit Margin'] = df['Profit'] / df['Sales']

# Group the data by 'Discount' and calculate average sales and profit for each discount level
discount_sales_profit = df.groupby('Discount')[['Sales', 'Profit']].mean().reset_index()

# Create a figure with dual y-axes
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Average Sales on the primary y-axis
color = 'skyblue'
ax1.set_xlabel('Discount Level')
ax1.set_ylabel('Average Sales', color=color)
ax1.plot(discount_sales_profit['Discount'], discount_sales_profit['Sales'], marker='o', label='Average Sales', color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a secondary y-axis for Average Profit
ax2 = ax1.twinx()
color = 'salmon'
ax2.set_ylabel('Average Profit', color=color)
ax2.plot(discount_sales_profit['Discount'], discount_sales_profit['Profit'], marker='o', label='Average Profit', color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Set title and labels
plt.title('Effect of Discount on Sales and Profit')
plt.xlabel('Discount Level')

# Display a legend for both lines
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
ax1.legend(lines, labels, loc='upper left')

# Show the dual-axis line chart
plt.tight_layout()
plt.grid(True)
plt.show()

# Output the analysis summary
print("\033[1mAnalysis Summary:\033[0m")
print(discount_sales_profit)
```

### 5.2 Are there optimal discount ranges for maximizing profit?

```python
# Calculate profit margin (Profit / Sales)
df['Profit Margin'] = df['Profit'] / df['Sales']

# Group the data by 'Discount' and calculate average profit for each discount level
discount_avg_profit = df.groupby('Discount')['Profit'].mean().reset_index()

# Create a scatter plot to visualize the relationship between discount and profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='Profit', data=discount_avg_profit, color='skyblue', label='Average Profit', s=100)

# Add a trendline (polynomial regression) to show the optimal discount range
z = np.polyfit(discount_avg_profit['Discount'], discount_avg_profit['Profit'], 2)
p = np.poly1d(z)
plt.plot(discount_avg_profit['Discount'], p(discount_avg_profit['Discount']), 'r--', label='Trendline')

# Set title and labels
plt.title('Optimal Discount Range for Maximizing Profit')
plt.xlabel('Discount Level')
plt.ylabel('Average Profit (USD)')

# Show legend
plt.legend()

# Show the scatter plot with trendline
plt.tight_layout()
plt.grid(True)
plt.show()

# Find the discount level corresponding to the maximum profit
optimal_discount = discount_avg_profit.loc[discount_avg_profit['Profit'].idxmax()]

# Output the analysis summary
print("\033[1mAnalysis Summary:\033[0m")
print("Optimal Discount Level for Maximizing Profit:")
print(optimal_discount)
```

## 6. Quantity Analysis:

```python
# Calculate profit margin (Profit / Sales)
df['Profit Margin'] = df['Profit'] / df['Sales']

# Create a scatter plot to visualize the relationship between quantity sold and profit
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantity', y='Profit', data=df, color='skyblue', label='Profit', s=100)

# Add a trendline (linear regression) to show the relationship
z = np.polyfit(df['Quantity'], df['Profit'], 1)
p = np.poly1d(z)
plt.plot(df['Quantity'], p(df['Quantity']), 'r--', label='Trendline')

# Set title and labels
plt.title('Relationship Between Quantity Sold and Profit')
plt.xlabel('Quantity Sold')
plt.ylabel('Profit (USD)')

# Show legend
plt.legend()

# Show the scatter plot with trendline
plt.tight_layout()
plt.grid(True)
plt.show()

# Calculate the correlation coefficient between quantity and profit
correlation_coefficient = df['Quantity'].corr(df['Profit'])

# Output the analysis summary
print("\033[1mAnalysis Summary:\033[0m")
print("Correlation Coefficient (Quantity vs. Profit):", correlation_coefficient)
```

## 8. Segmented Analysis:

### 8.1 Can customer segments be analyzed further to identify high-value customer behaviors and preferences?
